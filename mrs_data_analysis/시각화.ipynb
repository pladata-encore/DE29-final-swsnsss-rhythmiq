```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_excel('data/6.xlsx')

#    
word_freq = dict(zip(df['Tag'], df['Count']))

#  
wordcloud = WordCloud(font_path='data/NanumGothic.ttf', width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

#  
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
