```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_excel('data/카운터6.xlsx')

# 태그와 빈도수를 딕셔너리로 변환
word_freq = dict(zip(df['Tag'], df['Count']))

# 워드클라우드 생성
wordcloud = WordCloud(font_path='data/NanumGothic.ttf', width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# 워드클라우드 표시
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
